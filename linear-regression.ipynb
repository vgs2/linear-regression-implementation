# Linear Regression

### What is linear regression?

Linear regression is an algorithm used to find a line that best defines a set of data. By best defines, we can mean a lot of things, but in this case it will be the line which has the smallest mean square error related to the data.

To implement and show how it works we will be using the following libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
```

So, let≈õ suppose that we have the following data:


```python
X = np.array([139,126,90,144,163,136,61,62,41,120])
y = np.array([122,114,86,134,146,107,68,117,71,98])
plt.plot(X, y, 'bo')
```

Let's suppose that we will be receiving a new point and we only know its x, but not its y. It would be pretty handy in some situations to be able to predict it's y. That is where linear regression comes in handy. By having a line that represents the data nicely we can hopefully make a good prediction.

In this context, the error is the difference between the values predicted by the line and the real values. We need to find a line that minimizes this error.
If our line has the format $ y = a +bx $ 
We have that 
$ y_i = a + bx_i +e_i $ <br>
$ y_i - a - bx_i = e_i $ <br>
So, in order to find the minimun root square error, we need to find the values that minimizes the following formula: <br> 
$\sum_{i=1}^{n} {(y_i - a - bx_i)}^2 $ <br>
After getting both partial derivatives and some other simplifications, we get the following formulas: <br>

$b = \frac{{\sum_{i=1}^{n}{x_i(y_i-\overline{y})}}}{\sum_{i=1}^{n}{x_i(x_i - \overline{x})}}$ <br><br>
$ a = \overline{y} - b \overline{x} $


```python
def linear_regression(X, y):
    x_average = np.average(X)
    y_average = np.average(y)
    b = np.sum(X*(y-y_average)) / np.sum(X*(X-x_average))
    a = y_average - b*x_average
    return a, b
```

```python
linear_regression_X = np.array([np.min(X),np.max(X)])
linear_regression_y = a +b*linear_regression_X
plt.plot(linear_regression_X, linear_regression_y, 'r',X, y, 'bo')
```

As you can see in the plot above, the line fits pretty well in the data. But of course this is a pretty simple example.
